```python
# Import Libraries
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import pandas_datareader as web
from keras.models import Sequential
from keras.layers import Dense,LSTM
from sklearn.preprocessing import MinMaxScaler
plt.style.use('fivethirtyeight')
import datetime 
```

```python
import pandas as pd
import yfinance as yf
import datetime

# Define the time period for the data
start_date = datetime.datetime(2017, 1, 1)
end_date = datetime.datetime(2023, 7, 24)

# Get the stock data for AAPL from Yahoo Finance
symbol = 'AAPL'
df = yf.download(symbol, start=start_date, end=end_date)

# Show the dataset
print(df.head())
```

```python
# Get the size and shape of the datashape
df.shape
```

```python
# Visualize the closing prize of data
plt.figure(figsize=(12,8))
plt.title("Closing Prie of the AAPL Dataset")
plt.plot(df['Close'])
plt.xlabel('Date',fontsize=18)
plt.ylabel('Close Price USD($)',fontsize=18)
plt.show()         
         
```

```python
# Create the new dataframe with Close column.
data = df.filter(['Close'])
# Convert the dataframe into numpy array
dataset = data.values
# Gets the dataset to train the model on
training_data_len = math.ceil(len(dataset)*.8)
```

```python
training_data_len
```

```python
# Scale the data
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)
scaled_data
```

```python
# Creating the training datasets
# Creating the scaled datasets
train_data = scaled_data[0:training_data_len , :]
# Split the dataset into train and testdataset
X_train = []
y_train = []

for i in range(60,len(train_data)):
    X_train.append(train_data[i-60:i , 0])
    y_train.append(train_data[i,0])
    if i <=61:
        print(X_train)
        print(y_train)
        print()
    
```

```python
# COnvert the X_train,y_train into numpy array
X_train,y_train, =np.array(X_train),np.array(y_train)
```

```python
# Reshape the data as LSTM model required the 3d data
X_train = np.reshape(X_train,(X_train.shape[0],X_train.shape[1],1))
```

```python
X_train.shape
```

```python
# Build the LSTM model
model = Sequential()
model.add(LSTM(50,return_sequences=True,input_shape = (X_train.shape[1],1)))
model.add(LSTM(50,return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
# Complie the model
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
# Train the model
model.fit(X_train,y_train,batch_size=1,epochs=1)
```

```python
# Create the testing dataset
# Create a new array containing scale values from 1260 to 1648b
test_data = scaled_data[training_data_len -60 :, :]
# Create the dataset X_test and y_test
X_test = []
y_test = dataset[training_data_len:,:]
for i in range(60,len(test_data)):
    X_test.append(test_data[i-60:i,0])

```

```python
# Convert the data into numpy array
X_test = np.array(X_test)
```

```python
# Reshape the data into 3 dimension as LSTM required it in 3D
X_test = np.reshape(X_test,(X_test.shape[0],X_test.shape[1],1))
```

```python
# Get the model Predicted values
predictions = model.predict(X_test)
predictions=scaler.inverse_transform(predictions)
```

```python
# Evaluate the model by mean square error(RMSE)
rmse = np.sqrt(np.mean(predictions - y_test)**2)
rmse
```

```python
#Plot the data
train = data[:training_data_len]
valid = data[training_data_len:]
valid['Predictions'] = predictions
#Visualize the data
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date',fontsize = 16)
plt.ylabel('Closing Price ($)',fontsize=16)
plt.plot(train['Close'])
plt.plot(valid[['Close','Predictions']])
plt.legend(['Train','Val','Predictions'],loc='lower right')
plt.show()
```
